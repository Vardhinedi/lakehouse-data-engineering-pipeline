```python
# ============================================
# BRONZE LAYER PIPELINE
# ============================================

from pyspark.sql.functions import *

# ============================================
# CLEAN OLD BRONZE DATA
# ============================================

dbutils.fs.rm(
    "/Volumes/workspace/default/lakehouse_raw_data/bronze/orders",
    recurse=True
)

dbutils.fs.rm(
    "/Volumes/workspace/default/lakehouse_raw_data/bronze/customers",
    recurse=True
)

# ============================================
# READ CUSTOMERS JSON
# ============================================

customers_df = spark.read \
    .option("multiLine", "true") \
    .json(
        "/Volumes/workspace/default/lakehouse_raw_data/customers.json"
    )

# ============================================
# READ ORDERS JSON
# ============================================

orders_df = spark.read \
    .option("multiLine", "true") \
    .json(
        "/Volumes/workspace/default/lakehouse_raw_data/orders.json"
    )

# ============================================
# SHOW DATA
# ============================================

print("CUSTOMERS DATA")
display(customers_df.limit(20))

print("ORDERS DATA")
display(orders_df.limit(20))

# ============================================
# PRINT SCHEMA
# ============================================

customers_df.printSchema()
orders_df.printSchema()

# ============================================
# WRITE BRONZE TABLES
# ============================================

customers_df.write \
    .format("delta") \
    .mode("overwrite") \
    .save(
        "/Volumes/workspace/default/lakehouse_raw_data/bronze/customers"
    )

orders_df.write \
    .format("delta") \
    .mode("overwrite") \
    .save(
        "/Volumes/workspace/default/lakehouse_raw_data/bronze/orders"
    )

print("Bronze Layer Created Successfully!")
```
